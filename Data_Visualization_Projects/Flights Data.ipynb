```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv')
df.head()
```

```python
df_annual = df.groupby('year').sum(['passengers'])
df_annual.head()
```

```python
df_pivot = df.pivot(index='year',columns='month',values='passengers')
df_pivot.head()
```

```python
fig = plt.figure(figsize=(14,6))
ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

df_annual.plot(kind='line',ax=ax0)
ax0.set_title('Total Passenger by Year')
ax0.set_xlabel('Year')
ax0.set_ylabel('Passenger')

df_pivot.plot(kind='line',ax=ax1)
ax1.set_title('Total Passenger by Month')
ax1.set_xlabel('Years')
ax1.set_ylabel('Passenger')
plt.show()
```

```python
plt.figure(figsize=(14,6))
df_pivot.plot(kind='area',stacked=False,alpha=0.25)
plt.title('Total Passenger by Month')
plt.xlabel('Years')
plt.ylabel('Passenger')
plt.show()
```
