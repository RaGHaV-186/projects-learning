```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_mpg = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv")
df_mpg.head()
```

```python
df_mpg.info()
```

```python
df_mpg.boxplot(column="mpg", by="origin")
plt.title('Fuel Efficiency by Origin')
plt.ylabel('MPG')
plt.show()
```

```python
df_clean = df_mpg.dropna()

df_clean.plot(kind='scatter',x='horsepower',y='mpg')

plt.title("Horsepower vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")

plt.show()
```

```python
bubble_size = df_clean['weight'] / 20

df_clean.plot(kind='scatter', x='horsepower', y='mpg',
              s=bubble_size, alpha=0.5, figsize=(12, 6))

plt.title('Horsepower vs MPG (Bubble Size = Weight)')
plt.show()
```
