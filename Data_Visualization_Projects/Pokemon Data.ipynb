```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv')
df.head()
```

```python
df_water = df.loc[df['Type 1'] == "Water"]
df_water.head()
```

```python
df_water['Speed'].plot(kind='hist')
plt.title("Speed Distribution of Water Pokemon")
plt.show()
```

```python
df_types = df['Type 1'].value_counts()
df_types.head()
```

```python
df_types.plot(kind='barh')
plt.title("Types of Pokemon")
plt.show()
```

```python
df_legen = df['Legendary'].value_counts()
df_legen.head()
```

```python
df_legen.plot(kind='pie',autopct = '%1.1f%%')
plt.title("Legendary vs Non Legendary")
plt.show()
```
